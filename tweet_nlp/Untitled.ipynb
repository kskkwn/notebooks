```python
import pandas as pd
import numpy as np
import re
tweets = pd.read_csv("tweets/tweets.csv")
```

```python
tweets["source"].unique()
```

```python
tweets = tweets[tweets["in_reply_to_status_id"].isnull() & tweets["retweeted_status_id"].isnull()]["text"]
```

```python
def formatting(tweet):
    tweet = tweet.replace("\n", "")
    tweet = re.sub(r"http.*", "", tweet)
    tweet = re.sub(r"#.*", "", tweet)
    return tweet

tweets = list(map(formatting ,tweets.tolist()))
```

```python
import MeCab
m = MeCab.Tagger("-Owakati")
parsed = [m.parse(tweet).strip().split(" ") for tweet in tweets]
```
