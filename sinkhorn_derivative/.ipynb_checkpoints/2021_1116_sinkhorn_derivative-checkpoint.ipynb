```python
import numpy as np
import pylab as plt
from scipy.stats import norm
```

```python
import torch
import torch.nn

import ot
```

```python
X = torch.randn(10, 2)
Y = torch.randn(10, 2) + np.ones(2)*5
```

```python
plt.scatter(X[:,0], X[:,1])
plt.scatter(Y[:,0], Y[:,1])
```

```python
def softmin(z, epsilon):
    return -torch.logsumexp(-z/epsilon, dim=1)*epsilon
```

```python
n = X.shape[0]
m = Y.shape[0]
```

```python
epsilon = 0.001
```

```python
a = torch.ones(n)/n
b = torch.ones(m)/m
```

```python
f = torch.ones(n).double()
g = torch.zeros(m).double()
C = torch.sum((X[:, None] - Y[None])**2, dim=2)
C = C/C.max()

for i in range(500):
    S = C - f[:, None] - g[None]
    f = softmin(S, epsilon=epsilon) + f + epsilon * torch.log(a)
    S = C - f[:, None] - g[None]
    g = softmin(S.t(), epsilon=epsilon) + g + epsilon * torch.log(b)
```

```python
K = torch.exp(-C/epsilon)
P = torch.diag(torch.exp(f/epsilon)) @ K @ torch.diag(torch.exp(g/epsilon))
```

```python
plt.scatter(X[:,0], X[:,1])
plt.scatter(Y[:,0], Y[:,1])
for i,p_i in enumerate(P):
    for j,p_ij in enumerate(p_i):
        plt.plot([X[i, 0], Y[j,0]], [X[i, 1], Y[j,1]], alpha=float(p_ij.numpy())*2)
```

```python
plt.scatter(X[:,0], X[:,1])
plt.scatter(Y[:,0], Y[:,1])
for i,p_i in enumerate(ot.sinkhorn(M=C, a=a, b=b, reg=epsilon)):
    for j,p_ij in enumerate(p_i):
        plt.plot([X[i, 0], Y[j,0]], [X[i, 1], Y[j,1]], alpha=float(p_ij)*2)
```
