```python
import numpy as np
import pylab as plt
import matplotlib.gridspec as gridspec
```

```python
def plot_path(paths, A, B, C):
    gs = gridspec.GridSpec(2, 2,
                       width_ratios=[1,5],
                       height_ratios=[5,1]
                       )
    ax1 = plt.subplot(gs[0])
    ax2 = plt.subplot(gs[1])
    ax4 = plt.subplot(gs[3])

    ax2.pcolor(C, cmap=plt.cm.Blues)
    
    for path in paths:
        ax2.plot(path[:,0]+0.5, path[:,1]+0.5, c="C3")
    
    ax4.plot(A)
    ax1.invert_xaxis()
    ax1.plot(B, range(len(B)), c="C1")
    
    ax2.set_xlim(0, len(A))
    ax2.set_ylim(0, len(B))
    
    plt.show()
```

```python
T = 30
t = .4
x = np.sin(np.array(range(T*10)) / 10)
y = np.sin((np.array(range(T)) / 10 + t * np.pi))
```

```python
import numpy as np
from spring2 import partial_dtw
```

```python
cost, pathes = partial_dtw(x,y, 10)
C = (np.array(x).reshape(1, -1) - np.array(y).reshape(-1, 1))**2
plot_path(pathes, x,y, C)
```
